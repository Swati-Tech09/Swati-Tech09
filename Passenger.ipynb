```python
import os
import numpy as np
import pandas as pd
```

```python
os.getcwd()
os.chdir('D:\Data')
ds=pd.read_csv("Passenger.csv")
ds.head(10)
```

```python
ds.info()
```

```python
#Label Encoding to convert the Package into order preference
from sklearn.preprocessing import LabelEncoder
# creating instance of labelencoder
labelencoder = LabelEncoder()
#Assigning numerical values and storing in another column
ds['Class'] = labelencoder.fit_transform(ds['Package'])
ds
```

```python
from sklearn.preprocessing import OneHotEncoder
# creating instance of one-hot-encoder
One_enc = OneHotEncoder(handle_unknown='ignore')
#Passing gender variable
enc_df = pd.DataFrame(One_enc.fit_transform(ds[['Gender']]).toarray())
# merge with passenger dataset on key values
Enc_ds= ds.join(enc_df)
Enc_ds
```

```python
# generate binary values using get_dummies
dum_df = pd.get_dummies(ds['Destination'])
dum_df

```

```python
Passenger_Df=pd.concat([ds,dum_df],axis='columns')
Passenger_Df
```
